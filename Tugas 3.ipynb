# Analisa Data Ekspor di Indonesia berdasarkan negara tujuan.

```python
import pandas as pd
```

```python
data_ekspor = pd.read_csv('processed-nilai-ekspor-negara-tujuan-2005-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_ekspor
```

# Deskripsi Data 
Dataset ini berisi mengenai nilai ekspor Indonesia dari berbagai negara dalam rentang waktu 2005 - 2013. 
Data ini berisi informasi mengenai negara, benua, region tujuan ekspor serta nilai ekspornya.

Penjelasan mengenai Variabel pada Dataset ini:

- tahun: Tahun 
- bulan: Bulan
- benua: Nama Benua
- region: Nama Region
- nama_negara: Nama Negara tujuan
- nilai_ekspor: Nilai ekspor (dalam Ribu USD)

Catatan:

Pemekaran menjadi 27 negara sejak Bulgaria dan Romania bergabung dengan Uni Eropa pada Januari 2007
Untuk "Ekspor yang tidak dapat diklasifikasikan" terdiri dari barang yang diperoleh di pelabuhan oleh sarana pengangkut, barang untuk diperbaiki, dan penyesuaian cakupan ekspor industri Batam

-Sumber : https://data.go.id/dataset/nilai-ekspor-indonesia-berdasarkan-negara-tujuan

# Statistik tentang Data

```python
data_ekspor.columns
```

```python
data_ekspor.groupby('benua').sum()['nilai_ekspor']
```

```python
data_ekspor.groupby('region').sum()['nilai_ekspor']
```

```python
data_ekspor.groupby('nama_negara').sum()['nilai_ekspor']
```

```python
s = data_ekspor.nilai_ekspor.sum()
mi = data_ekspor.nilai_ekspor.min()
ma = data_ekspor.nilai_ekspor.max()
me = data_ekspor.nilai_ekspor.mean()
md = data_ekspor.nilai_ekspor.median()
mo = data_ekspor.nilai_ekspor.mode()

maa =data_ekspor.nama_negara.max()
mii =data_ekspor.nama_negara.min()

maa2= data_ekspor.region.max()
mii2= data_ekspor.benua.min()


maa1= data_ekspor.benua.max()
mii1= data_ekspor.benua.min()



print("Total nilai ekspor Indonesia berdasarkan negara tujuan sebesar\t:",s)
print("Rata-rata ekspor Indonesia berdasarkan negara tujuan\t\t:",me)
print("Median data ekspor Indonesia berdasarkan negara tujuan\t\t:",md)
print("Nilai ekspor tertinggi Indonesia berdasarkan negara adalah", maa, ", sebesar", ma)
print("Nilai ekspor terendah Indonesia berdasarkan negara adalah", mii, ", sebesar", mi)


print("Nilai ekspor tertinggi Indonesia berdasarkan Benua adalah", maa1)
print("Nilai ekspor terendah Indonesia berdasarkan Benua adalah", mii1)

print("Nilai ekspor tertinggi Indonesia berdasarkan Region adalah", maa2)
print("Nilai ekspor terendah Indonesia berdasarkan Region adalah", mii2)


```

# Visualisasi Sederhana dari Data

```python
ax = data_ekspor.groupby('nama_negara').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Berdasar Negara Tujuan", legend=False,figsize=(10,10), fontsize=10)
```

```python
ax = data_ekspor.groupby('benua').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Berdasar Negara Tujuan", legend=True,figsize=(10,10), fontsize=10)
```

```python
ax = data_ekspor.groupby('region').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Berdasar Negara Tujuan", legend=True,figsize=(10,10), fontsize=10)
```

```python
data_ekspor.groupby('tahun').sum()['nilai_ekspor']
```

```python
ax = data_ekspor.groupby('tahun').sum()['nilai_ekspor']
ax = ax.plot(kind='line', title ="Nilai ekspor Berdasar Tahun", legend=True, fontsize=10)
```
